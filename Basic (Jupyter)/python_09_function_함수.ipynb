도전! 실전코드

```python
def get_students(n):
  students = {}

  for i in range(n):
    name = input()
    score = int(input())
    students[name] = score # dict에 key값으로 새로운 score 추가
  return students

# 학생의 이름과 점수를 받은 것을 바탕으로 학점(등급) 매길게요.

def get_grade(score) :
    if score >= 90:
      return "A"
    elif score >= 80:
      return "B"
    elif score >= 70:
      return "C"
    elif score >= 70:
      return "D"
    else :
      return "F"

n = int(input())
students = get_students(n)

for student_name in students.keys():
   print(student_name + ": "+ get_grade(students[student_name])) # 말을 잇기 위해서는 +를 붙혀야한다.!
```
