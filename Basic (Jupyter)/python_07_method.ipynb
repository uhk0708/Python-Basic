sequence 의 indexing 과 slicing
- indexing : sequence에서 한 원소를 가져오는 것
- slixing : sequence에서 일부분을 가져오는 것

```python
list_4 = [1,2,3,4,5,6]

print(list_4[3:])
```

```python
print(list_4[:-2])
print(list_4[:])
print(list_4[1::2])
```
