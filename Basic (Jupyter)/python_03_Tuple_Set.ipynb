Tuple(튜플) 정의

```python
t1 = ()
t2 = (1,2,3,4,5)
t3 = (2,4, '판다' , '강아지') # 걸렉션 타입 (여러 개의 자료형 소유가능)
```

```python
# 인덱싱(indexing)
print(t2[0])
print(t2[-1])


# 인덱시을 통해 튜플 내부 데이터 접근 가능
# 파이썬에서 index는 0부터 시작
# 파이썬에서 index에 '-' rlghrk qnxdmaus , enldptjqnxj tntwkfmf tosek.(-1,-2,-3,...)
print()
print(t3[-2])
```

```python
# 슬라이싱(slicing)
# >>  튜플 안의 값의 일부만 가져오는 것

print(t2[2:])
print(t2[:4])
print(t2[1:3])
print(t2[::2]) # 간격(interval)이 2칸씩 이동
print(t2[::-1]) # 역순으로 -1 칸씩
```

튜플의 더하기, 곱하기

```python
t1 = (1,2,3)
t2 = (4,5,6)
t3 = t1+t2
```

```python
print(t3)
```

```python
t5 = t1 * 4
print (t5)
```

```python
t4 = (1,2,3,4)
# t4[-2] = 5
# tuple은 값의 임의 변경 불가 (immutable)
t4 = list(t4) # tuple >> list
t4[-2] = 5
print(t4)
print()
t4 = tuple(t4) # list >> tuple 변경
print(t4)
```

교집합, 차집합, 합집합

```python
t2 = set([1,2,3,4,5])
t3 = set([2,4,'판다', '강아지'])
```

```python
# 교집합
print(t2.intersection(t3))
print(t2 & t3)
```

```python
# 차집합
print(t2.difference(t3))
print(t2 - t3)
```

```python
# 합집합
print(t2.union(t3))
print(t2 | t3)
```
