```python
#  주민등록번호로  성별(남/여) 구분하기

number = int(input())

if number %2 == 0 :  # number 2로 나눈 뒤 나머지가 0이니깐
  print("여자입니다")
else :
  print("남자입니다")
```

```python
# 의류 사이즈 분류
#100에서 10단위로 커지는 사이즈에서
# 분류기준
# 1) size가 150이하 : 아동용
# 2) size가 190이상 : 맞춤옷
# 3) 그 외: 성인용
```

```python
size = int(input())

if size<=150:
  print('아동용')
elif size >= 190:
  print('맞춤옷')
else:
  print('성인용')
```

while
 - 무한 반복(반복 지옥)
 - 반드시 정지조건(pass, contunue , break)

```python
'''
1. break : while 반복문 밖으로 빠져나감
2. continue : 다음 소스 코드를 실행하지 않고 다음 단계로 넘어감
3. pass : 아무 것도 하지 않고 넘김
'''
```

```python
number = 0 # 초기화

while True:
   if number>4: # 정지조건
     break

     print(number)
   number = number + 1

   print(number)

# sum = 0 (더하기)
# mul = 1 (곱하기)

```

```python
# countinue

number = 0

while number < 6:
   number= number + 1
   if number > 4:
     continue

   print(number)


```
