클래스(Class)

```python
# 클래스 정의 (빵 굽는 틀)

class Car:  # 클래스는 항상 첫번째를 대문자로
  color = "" # 속성(필드) 초기화
  speed = 0

  # 메소드(기능) 정의

  def speedUp(self, value):
    self.speed += value

  def speedDown(self, value):
    self.speed -= value
```

```python
# 인스턴스(instance)

myCar1 = Car()
myCar1.color = '빨강'
myCar1.speed = 0
```

```python
myCar2 = Car()
myCar2.color = '파랑'
myCar2.speed = 0
```

```python
myCar3 = Car()
myCar3.color = '녹색'
myCar3.speed = 0
```

```python
myCar1.speedUp(30)
print("내 첫번째 차 색상은 %s이며, 현재 속도는 %d km/h 입니다."%(myCar1.color , myCar1.speed))
```

```python
myCar1.speedUp(50)
print("내 첫번째 차 색상은 %s이며, 현재 속도는 %d km/h 입니다."%(myCar1.color , myCar1.speed))
```

```python
myCar1.speedDown(60)
print("내 첫번째 차 색상은 %s이며, 현재 속도는 %d km/h 입니다."%(myCar1.color , myCar1.speed))
```

```python
myCar2.speedUp(50)
print("내 두번째 차 색상은 %s이며, 현재 속도는 %d km/h 입니다."%(myCar2.color , myCar2.speed))
```

인스턴스 변수와 클래스 변수

1. 클래스 변수와 인스턴스 변수 구별하는 법
-  self 붙으면 >> instance 변수
-  class 붙으면 >> class 변수

2. 클래스, 인스턴스 변수 접근
- 인스턴스 변수 접근: instance.instance 변수
- 클래스 변수 접근 : instance.class변수, Class.class 변수

```python
# 클래스 정의

class Car:
  speed = 0 # instance 변수
  count = 0 # class 변수

  # 생성자
  def __init__(self,speed):
    self.speed = speed
    Car.count += 1
```

```python
K8 = Car(30)
print("K8의 현재 속도는 %d km , 생상된 자동차는 총 %d 대 입니다."%(K8.speed, K8.count))
# instance.class 변수 접근
```

```python
K8 = Car(60)
print("K8의 현재 속도는 %d km , 생상된 자동차는 총 %d 대 입니다."%(K8.speed, K8.count))
# instance.class 변수 접근
```

클래스의 상속

```python
# 메소드의 오버라이딩 구현

# super class(부모 클래스) 선언

class Car:
  speed = 0 # instance 변수(속성)

  #메소드(기능) 구현
  def speedUp(self, value):
    self.speed += value
    print("현재 속도는(sper class):%d"%self.speed)
```

```python
# sub class(자식 클래스)

class Sedan(Car):
  def speedUp(self, value):
    self.speed += value

    if self.speed >= 110:
      self.speed = 110
    print("현재 속도는(서브클래스):%d"%self.speed)

class Truck(Car):
  pass

# 인스턴스 생성
porter = Truck()
genesis = Sedan()

print('트럭:')
porter.speedUp(50)

print('승용차:')
genesis.speedUp(150)
```

```python
# 메소드의 오버라이딩 구현 (심화 예제)

# super class(부모 클래스) 선언

class Car:
  speed = 0 # instance 변수(속성)

  #메소드(기능) 구현
  def speedUp(self, value):
    self.speed += value
    print("현재 속도는(sper class):%d"%self.speed)
```

```python
# sub class(자식 클래스) 선언

class Sedan(Car):
  seat = 0 # 좌석 수(속성)

  def getSeat(self, value): # 좌석수 알아보기(기능, 메소드)
    self.seat += value
    print("현재 좌석수: %d"%self.seat)


  def speedUp(self, value):
    self.speed += value

    if self.speed >= 110:
      self.speed = 110
    print("현재 속도는(서브클래스):%d"%self.speed)


class Truck(Car):
  load = 0 # 적재랑(속성, 필드)

  # 적재량 알아보는 기능(메소드)
  def getLoad(self, value):
    self.load += value
    print("현재 적재량은 : %d"%self.load)


#인스턴스 생성
porter = Truck()
genesis = Sedan()

print("트럭: ")
porter.speedUp(50)
porter.getLoad(300)

print("승용차: ")
genesis.speedUp(150)
genesis.getSeat(4)
```
